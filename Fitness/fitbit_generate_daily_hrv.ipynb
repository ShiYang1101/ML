```python
import pandas as pd
import glob
```

```python
dir_list = glob.glob('./Fitbit/ShiYangLee/Sleep/Daily Heart Rate Variability Summary -*.csv')
dfs = []
for file in dir_list:
    data = pd.read_csv(file)
    dfs.append(data)
temp = pd.concat(dfs, ignore_index = True)
temp.head()
```

```python
temp.set_index('timestamp', inplace = True)
temp = temp.sort_index()
temp
```

```python
temp.to_csv('./Fitbit/ShiYangLee/Sleep/daily_hrv_full.csv')
```
